# Create a data vizualization


You can either work with a partner in the class (max. 2 people) or individually.

1. Identify an interesting question relevant to any dataset seen in class, tutorials or assignments thus far.

2. Create one data vizualization that you think answers the question. Your vizualization should fill the screen, and have a large clear title. The title should communicate the key piece of information that you will be presenting.

3. You (and your partner, if it applies) will give a 5 minute presentation next tutorial (Wednesday, Feb. 13).


## 1. Presentation

Imagine you are giving a talk to answer the question. You have come to a point in the talk where you are going to use the vizualization.

If you present with a partner:

> **Speaker 1 says**: “Our first point is that _____________________________.”
>
> **Speaker 2 says**: “Our second point is that ___________________ .”
>
> (Summarize two main ideas communicated by your vizualization)
>
> **Speaker 1 says**, “Here we see ________________________________.”
>
> **Speaker 2 says**, “Here we see ________________________________.”
>
> (Summarize two interesting features of your vizualization)
>
> Speaker 1/2 asks for questions.


If you present alone:

> **Speaker says**: “The key point is that _____________________________.”
>
> (Summarize two main ideas communicated by your vizualization)
>
> **Speaker says**, “Here we see ________________________________.”
>
> (Summarize two interesting features of your vizualization)
>
> Speaker asks for questions.


## 2. Write-up

Write a short document (one page max.) explaining your choice of the question and visualization for this communication exercise. 


## 3. Reflection

After the presentation in class, reflect on your presentation.

(a) Identify the positive points about your presentation. What did you do well? 

(b) Idenfify areas of improvement about your presention. What could you have done better? What are the main aspects of public speaking that you think you should improve?

(c) What did you learn about communication through this exercise?
