## Tutorial objectives and relationship to Assignment \#3 

The goal of these two tutorials is to :
1. (as a presenter) gain experience in presenting results of on-going work to data scientists;
2. (as a presenter) gain experience in seeking feedback from an audience comprising of data science colleagues;
3. (as an audience) gain experience in providing critical, constructive feedback to colleagues presenting their work, and 
4. (as a presenter) answering questions from an expert audience. 

These critique sessions aim for each team to get regular, preliminary feedback on their work during the exploration and development process, which should ultimately result in a stronger A3 project. Further, by engaging in constructively critiquing other team's work, each individual should also gain insights into their own work on the project.
